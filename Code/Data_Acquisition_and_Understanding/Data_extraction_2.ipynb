```python
import pandas as pd

def get_income_data(RICs):
    result_list = []
    for RIC in RICs:
        # Retrieve Financial Data for FY0
        dfFin, err = ek.get_data(RIC, 'TR.TotalRevenue;TR.GrossProfit;TR.EBITDA;TR.EBIT;TR.NetIncomeBeforeTaxes;TR.NetIncome;TR.TtlCmnSharesOut',{'Period':'FY0','Scale':'6','curn':'GBP'})

        # Retrieve Financial Data for FY-1
        dfFin2, err = ek.get_data(RIC, 'TR.TotalRevenue;TR.GrossProfit;TR.EBITDA;TR.EBIT;TR.NetIncomeBeforeTaxes;TR.NetIncome;TR.TtlCmnSharesOut',{'Period':'FY-1','Scale':'6','curn':'GBP'})

        # Convert the selected rows to lists
        listFin = dfFin.iloc[0].tolist()
        listFin2 = dfFin2.iloc[0].tolist()

        # Create a dictionary with the lists
        dataDict2 = {'FY0': listFin, 'FY-1': listFin2 }

        # Create a DataFrame from the dictionary and set the index to column names
        dfIncome = pd.DataFrame(dataDict2, index=dfFin.columns)

        # Add a column called 'instrument' with the RIC value
        dfIncome['instrument'] = RIC

        result_list.append(dfIncome)
    
    # Concatenate the data frames in the result_list into a single DataFrame
    result_df = pd.concat(result_list)
    
    return result_df

# Assuming you have a list of RICs
RICs = ['LSEG.L', 'MCO']

# Call the function to get the financial data
income_data_1 = get_income_data(RICs)

# Print the final DataFrame
income_data_1
```

```python
## balance data 
import pandas as pd
import time

def get_balance_data(RICs):
    result_list = []
    for RIC in RICs:
        # Retrieve Balance Sheet Data for FY0
        dfBal, err = ek.get_data(RIC, 'TR.CashAndSTInvestments;TR.PropertyPlantEquipmentTotalNet;TR.TotalAssetsReported;TR.CurrentLiabilities;TR.TotalLongTermDebt;TR.TotalLiabilities;TR.TotalEquity',{'Period':'FY0','Scale':6,'Curn':'GBP'})
        
        # Add a time delay of 1 second between each loop iteration
        time.sleep(1)

        # Retrieve Balance Sheet Data for FY-1
        dfBal2, err = ek.get_data(RIC, 'TR.CashAndSTInvestments;TR.PropertyPlantEquipmentTotalNet;TR.TotalAssetsReported;TR.CurrentLiabilities;TR.TotalLongTermDebt;TR.TotalLiabilities;TR.TotalEquity',{'Period':'FY-1','Scale':6,'Curn':'GBP'})
        
        # Add a time delay of 1 second between each loop iteration
        time.sleep(10)

        # Create a dictionary with the data
        dataDict3 = {'FY0': list(dfBal.iloc[0, :]), 'FY-1': list(dfBal2.iloc[0, :]) }

        # Create a DataFrame from the dictionary and set the index to column names
        dfBalance = pd.DataFrame(dataDict3, index=dfBal.columns)

        # Add a column called 'instrument' with the RIC value
        dfBalance['instrument'] = RIC

        result_list.append(dfBalance)
    
    # Concatenate the data frames in the result_list into a single DataFrame
    result_df = pd.concat(result_list)
    
    return result_df

# Assuming you have a list of RICs
RICs = ['LSEG.L', 'MCO', 'MORN.O', 'SPGI.K', 'MSCI.K']

# Call the function to get the balance sheet data
balance_data = get_balance_data(RICs)

# Print the final DataFrame
balance_data
```
